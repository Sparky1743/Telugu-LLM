```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
URL = "https://www.eenadu.net/telugu-news/general/daily-horoscope-for-may-10th-2024-friday-in-telugu/0600/124089546"
```

```python
r = requests.get(URL)
soup = BeautifulSoup(r.text, "html.parser")
# results = soup.find_all(['h1', 'p'])
def is_telugu(text):
    # Telugu Unicode range
    telugu_range = r'[\u0C00-\u0C7F]+'
    return re.match(telugu_range, text)
```

```python
# text = [result.text for result in results]
# article = ' '.join(text)

all_text = soup.get_text(separator=' ')
telugu_text = [text.strip() for text in all_text.split() if is_telugu(text)]
telugu_article = ' '.join(telugu_text)
```

```python
print(telugu_article)
```

```python
with open('telugusummary.txt', 'w', encoding='utf-8') as f:
    f.write(telugu_article)
```
